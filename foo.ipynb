Exercise 1. Solving Calcone

For similar problems, we check the boundary points values first. If it's a $C^{1}$ function, the derivative at local extrema should be 0. <br> Hence we proceed by finding the first and second derivatives of the original function, and use Newton's iterative method for finding 0s of the first derivative.

However if the initial point of iteration is not properly chosen, or 'close enough', then Newton's method could 'overshoot' and fail to converge to the closest zeros.

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x**2 + np.cos(x))**(2) - 10*(np.sin(5*x))
```

```python
def fp(x):
    return 2*(x**2 + np.cos(x))*(2*x - np.sin(x)) - 50*(np.cos(5*x))
```

```python
def fpp(x):
    return 2*(2*x - np.sin(x))*(2*x - np.sin(x)) \
           + 2*(x**2 + np.cos(x))*(2 - np.cos(x)) \
           + 250*(np.sin(5*x))
```

```python
rr = np.arange(0,2,0.000001)
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.xticks(np.arange(min(rr), max(rr)+1, 0.1))
plt.plot(rr,f(rr))
plt.show()
```

```python
a = [0.25]
tol = 1e-16
for k in range(1,20):
    k = 1
    fpi = fp(a[k-1])
    fppi = fpp(a[k-1])
    a.append(a[k-1] - (fpi/fppi))
    if (abs(fpi) <= tol):
        nfinal = k
        break
    k = k + 1
    
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.xticks(np.arange(0, 20, 1))
plt.yticks(np.arange(0.2, 0.32, 0.006))
plt.plot(a)
plt.show()
```
